#***SASyHR Prototype Model***

1. $S$: Susceptible
2. $A$: Assymptomatic
3. $S_y$: Symptomatic
4. $H$: Hospitalised
5. $R$: Recovered

### ***Differential Equations***

### $\frac{dS}{dt}= -[\beta_{A}\frac{A}{N}S + \beta_{S_y}\frac{S_y}{N}S]$
### $\frac{dA}{dt}= \beta_{A}\frac{A}{N}S-\alpha A -\epsilon_A A$
### $\frac{dS_y}{dt}= \beta_{S_y}\frac{S_y}{N}S + \alpha A -\gamma_{S_y}S_y - \epsilon_{s_y}S_y $
### $\frac{dH}{dt}= \gamma_{S_y}S_y - \rho_h H$
### $\frac{dR}{dt}=  \rho_h H + \epsilon_A A + \epsilon_{s_y}S_y $

## **Model 1** ##
### Parameters - $\beta_{A}=3/4$, $\beta_{S_y}=2.5/6.5$,   $a=\frac{1}{10}$,$\gamma_{S_y} = \frac{1}{5}$, $\epsilon_A=\frac{1}{4}$, $\epsilon_{S_y}=\frac{1}{6.5}$, $\rho_h = 0.6$  ,$N=1000$, $A=15$, $Sy=5$ ###

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sir_odes(t, x, b_a, b_sy,a, g_sy, e_a, e_sy, p_h):
    S = x[0]
    A = x[1]
    Sy = x[2]
    H = x[3]
    R = x[4]

    N = np.sum([S, A, Sy, H, R])

    dSdt = -(b_a/N)*A*S -(b_sy/N)*Sy*S
    dAdt = (b_a*A*S)/N - e_a*A - a*A
    dSydt = (b_sy*Sy*S)/N + a*A - g_sy * Sy - e_sy * Sy
    dHdt = g_sy * Sy - p_h * H
    dRdt = p_h* H + e_sy*Sy + e_a*A

    return [dSdt, dAdt, dSydt, dHdt, dRdt]

t_span = np.array([0, 100])  # Interval of integration (time limits)
t = np.linspace(t_span[0], t_span[1], t_span[1] + 1)  # Values of time
x_0 = [980, 15, 5,0, 0]  # Initial conditions for model variables: S, A, Sy, R
args = ((3/4), (2.5/6.5),1/10 ,(1/5), (1/4), (1/6.5), 0.6)  # Arguments for our model parameters: b_a, b_sy,a, g_sy, e_a, e_sy, p_h

solutions = solve_ivp(sir_odes, t_span, x_0, args=args, t_eval=t)
S = solutions.y[0]
A = solutions.y[1]
Sy = solutions.y[2]
H = solutions.y[3]
R = solutions.y[4]
N = S+A+Sy+H +R

plt.plot(t, S/N, label='Susceptible', color='green')
plt.plot(t, A/N, label='Assymptomatic', color='orange')
plt.plot(t, Sy/N, label='Symptomatic', color='red')
plt.plot(t, H/N, label='Hospitalised', color='black')
plt.plot(t, R/N, label='Recovered', color='blue')
plt.xlabel('Time [days]')
plt.ylabel('Proportion of Population')
plt.legend()
plt.show()

print('S[-1]=', round(S[-1], 0))
print('A[-1]=', round(A[-1], 0))
print('Sy[-1]=', round(Sy[-1], 0))
print('H[-1]=', round(A[-1], 0))
print('R[-1]=', round(R[-1], 0))
```
