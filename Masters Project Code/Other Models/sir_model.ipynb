### ***Differential Equations***

### $\frac{dS}{dt}= \frac{- \beta SI}{N}$
### $\frac{dI}{dt}= \frac{\beta IS}{N} - \gamma I$
### $\frac{dR}{dt}= \gamma I$

\

---- 
## **Model 1** ##
### Parameters - $R_0=2.5$, $\gamma=6.5$, $N=1000$, $I=5$ ###




```python
#@title
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def sir_odes(x, t, b, g, N):
    S = x[0]
    I = x[1]
    R = x[2]

    dSdt = -(b/N)*S*I
    dIdt = (b/N)*S*I - g*I
    dRdt = g*I

    return [dSdt, dIdt, dRdt]

t = np.linspace(0, 2, 1000)  # Time series
x_0 = [995, 5, 0]  # Initial conditions for model variables: S, I, R respectively
args = (16.25, 6.5, 1000)  # Arguments for our model parameters: \beta, \gamma, N

x = odeint(sir_odes, x_0, t, args)

S = x[:,0]
I = x[:,1]
R = x[:,2]

t = t*500
plt.plot(t, S/args[2], label='Susceptible')
plt.plot(t, I/args[2], label='Infected')
plt.plot(t, R/args[2], label='Recovered')
plt.xlabel('Time [days]')
plt.ylabel('Proportion of Population')
plt.legend()
plt.show()
```

## **Model 4** ##
### Parameters - $R_0=2$, $\gamma=0.1$, $N=1000$, $I=1$ ###

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sir_odes(t, x, b, g, N):
    S = x[0]
    I = x[1]
    R = x[2]

    dSdt = -(b/N)*S*I
    dIdt = (b/N)*S*I - g*I
    dRdt = g*I

    return [dSdt, dIdt, dRdt]

t_span = np.array([0,200]) # Time limits
t = np.linspace(t_span[0], t_span[1], t_span[1] + 1)  # Time series
x_0 = np.array([999, 1, 0])  # Initial conditions for model variables: S, I, R respectively
args = (0.2, 0.1, np.sum(x_0))  # Arguments for our model parameters: \beta, \gamma, N

sol = solve_ivp(sir_odes, t_span, x_0, args = args, t_eval = t)

R_0 = args[0]/args[1]
S = sol.y[0]
I = sol.y[1]
R = sol.y[2]
N = S + I + R

plt.plot(t, S, label='Susceptible')
plt.plot(t, I, label='Infected')
plt.plot(t, R, label='Recovered')
plt.plot(t, N, label ='Total Population')
plt.xlabel('Time [days]')
plt.ylabel('Population')
plt.legend()
plt.show()
print("R_0 is:", R_0)
print("End-population:", int(round(N[-1],0)))
```

## **Model 5: Algeria** ##
### Parameters - $R_0=1.25$, $\gamma=0.0455$, $N=45,000,000$, $I=100$ ###



```python
#@title
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def sir_odes(t, x, b, g, N):
    S = x[0]
    I = x[1]
    R = x[2]

    dSdt = -(b/N)*S*I
    dIdt = (b/N)*S*I - g*I
    dRdt = g*I

    return [dSdt, dIdt, dRdt]

t_span = np.array([0,2000]) # Time limits
t = np.linspace(t_span[0], t_span[1], t_span[1] + 1)  # Time series
x_0 = np.array([44999900, 100, 0])  # Initial conditions for model variables: S, I, R respectively
args = (0.056, 0.0455, np.sum(x_0))  # Arguments for our model parameters: \beta, \gamma, N

sol = solve_ivp(sir_odes, t_span, x_0, args=args, t_eval = t)

R_0 = args[0]/args[1]
S = sol.y[0]
I = sol.y[1]
R = sol.y[2]
N = S + I + R

plt.plot(t, S, label='Susceptible')
plt.plot(t, I, label='Infected')
plt.plot(t, R, label='Recovered')
plt.plot(t, N, label ='Population')
plt.xlabel('Time [days]')
plt.ylabel('Number')
plt.legend()
plt.show()
print("R_0 is:", round(R_0, 3))
print("End-population:", int(round(N[-1],0)))
```
